#Popularity based Recommendation System


```python
#We are going to recommend top 50 books having highest average ratings but there should be more than 250 ratings over them
ratings_with_name = ratings.merge(books,on='ISBN')
ratings_with_name
```

```python
num_rating_df=ratings_with_name.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_rating_df.rename(columns={'Book-Rating':'num_ratings'},inplace=True)
num_rating_df
```

```python
avg_rating_df=ratings_with_name.groupby('Book-Title')['Book-Rating'].mean().reset_index()
avg_rating_df.rename(columns={'Book-Rating':'avg_ratings'},inplace=True)
avg_rating_df
```

```python
popularity_df=num_rating_df.merge(avg_rating_df,on='Book-Title')
popularity_df
```

```python
popularity_df=popularity_df[popularity_df['num_ratings']>=250].sort_values('avg_ratings',ascending=False).head(50)
```

```python
popularity_df
```

```python
popularity_df.merge(books,on='Book-Title').drop_duplicates('Book-Title')[['Book-Title','Book-Author','Image-URL-M','num_ratings','avg_ratings']].reset_index()
```

#Collaborative Filtering Based Recommendation System

here we will try to create a dataframe where in the columns there will be all the users and in rows there will be all the books with the rating given by particular user.

We are only going to consider those users who had rated atleast 200 which shows that those users are knowledgeble and reads a lot of books.Also only those books will be considered that are having atleast 50 ratings which shows their popularity.

```python
x=ratings_with_name.groupby('User-ID').count()['Book-Rating']>200
padhe_likhe_users=x[x].index
```

```python
ratings_with_name['User-ID'].isin(padhe_likhe_users)
```

```python
filtered_ratings=ratings_with_name[ratings_with_name['User-ID'].isin(padhe_likhe_users)]
```

```python
y=filtered_ratings.groupby('Book-Title').count()['Book-Rating']>=50
famous_books=y[y].index
```

```python
final_ratings=filtered_ratings[filtered_ratings['Book-Title'].isin(famous_books)]
```

```python
final_ratings.drop_duplicates()
```

```python
final_ratings
```

```python
pt=final_ratings.pivot_table(index='Book-Title',columns='User-ID',values='Book-Rating')
```

```python
pt
```

```python
pt.fillna(0,inplace=True)
```

```python
from sklearn.metrics.pairwise import cosine_similarity
```

```python
similarity_score=cosine_similarity(pt)
```

```python
similarity_score[0]
```

```python
similarity_score.shape
```

Now we have to create a functionnamed recommend that will take name of a book as input and will return the names of 5 books in the recommendation.


```python
sorted(list(enumerate(similarity_score[0])),key= lambda x:x[1],reverse=True)[1:6]
```

```python
def recommend(book_name):
  # index fetch
  index= np.where(pt.index == book_name )[0][0]
  distance=similarity_score[index]
  similar_items=sorted(list(enumerate(similarity_score[index])),key= lambda x:x[1],reverse=True)[1:11]

  for i in similar_items:
    print(pt.index[i[0]])

```
